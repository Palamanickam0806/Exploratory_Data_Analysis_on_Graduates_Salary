Non visual data analysis

```python
for column in numeric_column:
  print('stat of ',column)
  print('mean :' , df[column].mean())
  print('minimum :' , df[column].min())
  print('maximum :' , df[column].max())
  print('standard deviation :' , df[column].std())
  print('skewness :' , df[column].skew())
  print('*'*10,'end','*'*20)
  print()
```

```python
for i in numeric_column:
  fig,axes = plt.subplots(1,3,figsize=(20,8))
  plt.tight_layout(pad=2)

  sns.boxplot(df[i],ax=axes[0],fill=True)
  axes[0].set_title(f"Box plot - '{i}'")
  axes[0].set_xlabel(i)

  sns.kdeplot(df[i],fill=True,ax=axes[1])
  axes[1].set_title(f"Probability density function plot - '{i}'")
  axes[1].set_xlabel(i)
  axes[1].set_ylabel('density')


  sns.histplot(df[i],fill=True,ax=axes[2])
  axes[2].set_title(f"Histogram plot - '{i}'")
  axes[2].set_xlabel(i)
  axes[2].set_ylabel('frequency')
  plt.show

```

# **Univariate Categorical Column Analysis**

```python
categorical_column = ['Gender','10board','12board','Degree','Specialization','CollegeState']
for i in categorical_column:
  print('*'*20,i,20*'*')
  print(f'Value count of {i}:::')
  print(df[i].value_counts())
  print()
```

```python
viz_categorical_column = ['Gender','Degree']
for i in viz_categorical_column:
    plt.figure(figsize=(6, 6))
    plt.pie(df[i].value_counts(), labels=df[i].value_counts().index, autopct='%1.1f%%')
    plt.title(f"Pie Chart of '{i}'")
    plt.show()
```

```python
viz_categorical_column = ['Specialization','CollegeState']
for i in viz_categorical_column:
    plt.figure(figsize=(20, 10))
    sns.countplot(df[i],palette='Set2',legend=False)
    plt.title(f"Count Plot of '{i}'")
    plt.show()
```

```python
today = pd.to_datetime('today')
df['age'] = today.year - df['DOB'].dt.year
plt.title(f"Probability distribution function plot of age")
sns.kdeplot(df['age'],fill=True)
plt.show()

plt.pie(df['age'].value_counts(), labels=df['age'].value_counts().index, autopct='%1.1f%%')
plt.title(f"Pie Chart of age")
```

## **Bivariate analysis**

```python
numeric = ['10percentage','12percentage','collegeGPA','CollegeTier','GraduationYear','age']
for j, i in enumerate(numeric, 1):
  plt.subplot(3,2, j)
  sns.scatterplot(x=i,y='Salary',data=df)
  plt.title(f"Scatter Plot: {i} vs Salary")
  plt.xlabel(i)
  plt.ylabel("Salary")
  plt.grid(True)
plt.tight_layout()
plt.show
```

```python
numeric = ['10percentage','12percentage','collegeGPA','CollegeTier','GraduationYear','age']
for j, i in enumerate(numeric, 1):
  plt.subplot(3,2, j)
  sns.lineplot(x=i,y='Salary',data=df)
  plt.title(f"Scatter Plot: {i} vs Salary")
  plt.xlabel(i)
  plt.ylabel("Salary")
  plt.grid(True)
plt.tight_layout()
plt.show()
```
